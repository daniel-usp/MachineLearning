````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dados_imoveis = pd.read_excel('dados_imoÌveis_SP.xlsx')

dados_imoveis.dropna(inplace=True)

# 1. RelaÃ§Ã£o entre Valor Total e Ãrea Ãštil dos ImÃ³veis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dados_imoveis, x='area_util', y='valor_total')
plt.title('RelaÃ§Ã£o entre Valor Total e Ãrea Ãštil dos ImÃ³veis')
plt.xlabel('Ãrea Ãštil (mÂ²)')
plt.ylabel('Valor Total (R$)')
plt.show()

# 2. DistribuiÃ§Ã£o de ImÃ³veis por Quantidade de Quartos
plt.figure(figsize=(10, 6))
sns.histplot(dados_imoveis['quartos'], bins=10, kde=True)
plt.title('DistribuiÃ§Ã£o de ImÃ³veis por Quantidade de Quartos')
plt.xlabel('Quantidade de Quartos')
plt.ylabel('FrequÃªncia')
plt.show()

# 3. Impacto do NÃºmero de Vagas na ValorizaÃ§Ã£o do ImÃ³vel
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dados_imoveis, x='vagas', y='valor_total')
plt.title('Impacto do NÃºmero de Vagas na ValorizaÃ§Ã£o do ImÃ³vel')
plt.xlabel('NÃºmero de Vagas')
plt.ylabel('Valor Total (R$)')
plt.show()

# 4. ComparaÃ§Ã£o dos PreÃ§os MÃ©dios dos ImÃ³veis por Bairro
media_por_bairro = dados_imoveis.groupby('bairro')['valor_total'].mean().sort_values(ascending=False)
plt.figure(figsize=(20, 6))
media_por_bairro.plot(kind='bar')
plt.title('PreÃ§os MÃ©dios dos ImÃ³veis por Bairro')
plt.xlabel('Bairro')
plt.ylabel('PreÃ§o MÃ©dio (R$)')
plt.xticks(rotation=90)
plt.show()

# 5. InfluÃªncia da Proximidade com EstaÃ§Ãµes de MetrÃ´ no Valor dos ImÃ³veis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dados_imoveis, x='dist', y='valor_total')
plt.title('InfluÃªncia da Proximidade com EstaÃ§Ãµes de MetrÃ´ no Valor dos ImÃ³veis')
plt.xlabel('DistÃ¢ncia atÃ© EstaÃ§Ã£o de MetrÃ´ (m)')
plt.ylabel('Valor Total (R$)')
plt.show()

````
